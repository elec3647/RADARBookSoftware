# Probability of Detection with Rayleigh Noise
***

Referring to Section 6.1.3, for coherent detectors, illustrated in Figure 5.11, the noise is modeled  at the output of the in-phase and quadrature channels as Gaussian, with zero mean and noise power of $\sigma^2/2$.  The magnitude of the in-phase and quadrature channels is (Equation 6.12)

$$
    z = \sqrt{X_I^2 + X_Q^2},
$$

then the resulting probability density function follows a Rayleigh distribution given by (Equation 6.13)

$$
    p(z) = \frac{z}{\sigma^2}\, \exp\left(\frac{-z^2}{2\sigma^2}\right).
$$

Following the procedure in  Section 6.1.2, the probability of false alarm for the coherent detector is written as (Equation 6.14)


$$
    P_{{fa}} = \int\limits_{V_T}^\infty \frac{z}{\sigma^2}\, \exp\left(-\frac{z^2}{2\sigma^2}\right)\, dz = \exp\left(-\frac{V_T^2}{2\sigma^2}\right).
$$

Solving for the detection threshold gives (Equation 6.15)

$$
    V_T = \sqrt{-2\sigma^2 \ln{(P_{{fa}}})}.
$$
***

Begin by getting the library path

```python
import lib_path
```

Set the start and end signal to noise ratio (dB)

```python
snr_start = 0.0

snr_end = 20.0
```

Create the signal to noise array using the `linspace` routine from `scipy`

```python
from numpy import linspace

snr = 10.0 ** (linspace(snr_start, snr_end, 200) / 10.0)
```

Set the probability of false alarm

```python
pfa = 1e-6
```

Calcualte the probability of detection

```python
from Libs.detection.single_pulse import pd_rayleigh

pd = pd_rayleigh(snr, pfa)
```

Display the probability of detection for Rayleigh noise using the `matplotlib` routines

```python
from matplotlib import pyplot as plt

from numpy import log10



# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)



# Display the results

plt.plot(10.0 * log10(snr), pd, '')



# Set the plot title and labels

plt.title('Pd - Rayleigh Noise', size=14)

plt.xlabel('Signal to Noise (dB)', size=12)

plt.ylabel('Probability of Detection', size=12)



# Set the tick label size

plt.tick_params(labelsize=12)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)
```
